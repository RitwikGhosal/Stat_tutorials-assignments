### loading the rain data and counting the number of data points less than mean

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
os.listdir() # to check if the the file is there in your directory or not

df1 = pd.read_csv("annual_max_rainfall_1901_2014_cru_india_1x1 .xlsx - Sheet1.csv")
df1.plot(x = 'year',y = ['X80.518.5','X82.522.5'])
df1.describe() # describing the data

#-------- counting the number of data points less than mean value-------------------------
mean_num = df1['X80.518.5'].mean()
df2 = df1[df1['X80.518.5'] < mean_num]
data_points_1 = df2.shape[0]

mean_num_2 = df1['X82.522.5'].mean()
df3 = df1[df1['X82.522.5'] < mean_num_2]
data_points_2 = df3.shape[0]

print("number of data points less than mean for X80.518.5 is {} and for X82.522.5 is {}".format(data_points_1,data_points_2))
```

### Fitting a linear regression model using least squares

```python
df7 = pd.read_excel("data.xlsx")
#df7.plot(x = 'Duration', y = 'Volume')
plt.scatter(x = df7['Duration'],y = df7['Volume'])
X_mean = np.mean(df7['Duration'])
Y_mean = np.mean(df7['Volume'])
X = df7['Duration']
Y = df7['Volume']
n = 0
d = 0
for i in range(len(X)):
    n += (X[i] - X_mean)*(Y[i] - Y_mean)
    d += (X[i] - X_mean)**2
m = n / d
c = Y_mean - m*X_mean

print('The value of Beta_1 is {} and Beta_0 is {}'.format(m, c))
Y_pred = m*X + c # m is beta_1 and c is beta_0

plt.scatter(X, Y) # actual
# plt.scatter(X, Y_pred, color='red')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red') # predicted
plt.show()
print( 'The predicted value for 120 is {}'.format(m*120 + c))
```
